# 1_Basic data exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics



# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
rows , columns = df.shape
print(f'number of rows in dataset : {rows}')
print(f'number of columns in dataset : {columns}')

for column in df.columns:
    data_type = df[column].dtype
    print(f"Column : {column}")
    print(f"Datatype : {data_type}")
```

```python
for column in df.columns:
  nullsum = df[column].isnull().sum()
  print(f"Column {column} has {nullsum} missing values")
```

# 2_Descriptive Statistics

```python
for column in df.columns:
  if(df[column].dtype == 'int64' or df[column].dtype == 'float64'):
    test = df[column].interpolate(method= 'linear')
    max = df[column].max()
    min = df[column].min()
    mean = statistics.mean(test)
    median = statistics.median(test)
    mode = statistics.mode(test)
    standard_deviation = np.std(df[column])
    print(f"Column : {column}")
    print(f"Maximum value : {max}")
    print(f"Minimum value : {min}")
    print(f"Mean : {mean}")
    print(f"Median : {median}")
    print(f"Mode : {mode}")
    print(f"Standard deviaton : {standard_deviation}")


minimum_salary = df['TotalPay'].min()
maximum_salary = df['TotalPay'].max()
print(f"Minimum salary : {minimum_salary}")
print(f"Maximum salary : {maximum_salary}")
print(f"Salary range is : {minimum_salary} -> {maximum_salary}")







```

# 3_Data cleaning

```python
#during data exploration process , there were two columns with NaN value at every row of them , so it was better to remove them

df1 = df.drop('Notes' , axis=1)
df1 = df1.drop('Status' , axis= 1)

#I filled 609 cells in 'BasePay' column with 0 value
#I filled 4 cells in 'OvertimePay' column with 0 value
#I filled 4 cells in 'OtherPay' column with 0 value
#I filled 36163 cells in 'Benefits' column with 0 value
#I choosed 0 value because it will not affect the data record (NaN value mean 0 extra on salary)

df1['Benefits'].fillna(0 , inplace = True)
df1['BasePay'].fillna(0 , inplace = True)
df1['OtherPay'].fillna(0 , inplace = True)
df1['OvertimePay'].fillna(0 , inplace = True)
df1['Benefits'].fillna(0 , inplace = True)
df1
```

# 4_Data visualiazation

Salaries distribution

```python


bins=[i for i in range(0 , 600001,50000)]
# print(bins)
data =df['TotalPay']
plt.hist(data,bins=bins,edgecolor='black')
plt.title('Salary distribution')
plt.xlabel('Salary')
plt.ylabel('frequency')
plt.show()
```

```python
#Here ,The pie chart displays 10 job titles that has the highest amount of employees

jobfrequency = df['JobTitle'].value_counts()
jobtitle = df['JobTitle'][:10]
jf = jobfrequency[:10]
jobsum = sum(jobfrequency)
jfpercentage=(jf / jobsum)*100


plt.pie(jfpercentage , labels = jobtitle, autopct ="%1.1f%%")
plt.show()

```
